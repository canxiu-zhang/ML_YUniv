```python
from pycaret.classification import *
import pandas as pd
from sklearn.model_selection import train_test_split

lda_df_final = pd.read_csv('lda_df_final.csv')
lda_train, lda_test = train_test_split(lda_df_final, test_size=0.2, random_state=42)
```

```python
classification = setup(data = lda_train,  
                 target = 'encoded_sentiment',
                train_size = 0.8,
                fold_strategy = 'kfold',
                ignore_features=['UserName', 'ScreenName', 'Location', 'TweetAt', 'OriginalTweet', 'Topic_0', 'Topic_1', 'Topic_2','Topic_3'],
                normalize = True,
                 normalize_method = 'robust')
```

```python
best_model = compare_models()
```

```python
# from pycaret.classification import *
lr = create_model('lr')
```

```python
plot_model(lr, plot = 'confusion_matrix')
```

```python
pred_unseen = predict_model(lr, data = lda_test)
```

```python
save_model(lr, 'lr_final', verbose=True)
```
