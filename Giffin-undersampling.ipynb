```python
import numpy as np
from pycaret.classification import *
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
ru = RandomUnderSampler(random_state=123)
```

```python
rf_clf = setup(data=df, target="Bankrupt?", fix_imbalance=True, fix_imbalance_method=ru)
```

```python
models()
```

```python
rf = create_model('rf')
print(rf)
```

```python
plot_model(rf, plot='auc')
```

```python
plot_model(rf, plot='confusion_matrix', plot_kwargs={'percent': True})
```
